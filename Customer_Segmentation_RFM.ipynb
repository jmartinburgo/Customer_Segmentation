```python
#Cantidad de productos vendidos

df.groupby('Description').agg({'Quantity':'sum'}).sort_values(by='Quantity',ascending=False)
```

```python
#Total por factura

df.groupby("Invoice").agg({"TotalPrice": "sum"}).sort_values(by='TotalPrice',ascending=False)
```

```python
df.describe().T
```

```python
#Le asignamos el precio de los cuartiles 1% y 99% a los outliears

o1=df['Price'].quantile(.01)
o3=df['Price'].quantile(.99)
inter_o=o3-o1
up_v=o3+inter_o*1.5
low_v=o1-inter_o*1.5
low_v= low_v if low_v>=0 else 0

df.loc[df['Price']>up_v,'Price']=up_v
df.loc[df['Price']<low_v,'Price']=low_v

print(f'Percentil 1%: {o1}')
print(f'Percentil 99%: {o3}')
print(f'Rango Interpercentil: {inter_o}')
print(f'Limite Inferior: {low_v}')
print(f'Limite Superior: {up_v}')
df
```

```python
#Le asignamos las cantidades de los cuartiles 1% y 99% a los outliears

o1=df['Quantity'].quantile(.01)
o3=df['Quantity'].quantile(.99)
inter_o=o3-o1
up_v=o3+inter_o*1.5
low_v=o1-inter_o*1.5
low_v= low_v if low_v>=0 else 0

df.loc[df['Quantity']>up_v,'Quantity']=up_v
df.loc[df['Quantity']<low_v,'Quantity']=low_v

print(f'Percentil 1%: {o1}')
print(f'Percentil 99%: {o3}')
print(f'Rango Interpercentil: {inter_o}')
print(f'Limite Inferior: {low_v}')
print(f'Limite Superior: {up_v}')
df
```

```python
#Modificamos el formato de la columna de fecha

df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])
df
```
