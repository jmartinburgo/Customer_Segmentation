# **CRECION DE TABLA PARA APLICAR LOS MODELOS CLTV PREDICTION Y RFM**

```python
#Armamos la tabla de clientes con:
# - El periodo entre la ultima y la primera compra
# - El periodo entre hoy y la ultima compra
# - El periodo entre hoy y la primera compra
# - Cantidad total de compras
# - Monto total de compra

cltv_c=df.groupby('Customer ID').agg({'InvoiceDate':[lambda x:(today_date-x.max()).days,lambda x:(today_date-x.min()).days],
                                      'Invoice': lambda x : x.nunique(),
                                      'TotalPrice': lambda x : x.sum()
                                      })


cltv_c.head()
```

```python
#Eliminamos la primera fila con de nombres de colummnas

cltv_c.columns=cltv_c.columns.droplevel(0)

cltv_c.head()
```

```python
#Asignamos los nombres a las columnas
cltv_c.columns=["recency","T","frequency","monetary"]

cltv_c
```

```python
#Pasamos los dias a semanas y sacamos los que tiene frecuencia 0 y modificamos la columna monetary teniendo en cuenta los montos por compra

cltv_c["recency"]=cltv_c["recency"]/7
cltv_c["T"]=cltv_c["T"]/7
cltv_c=cltv_c[cltv_c["frequency"]>1]
cltv_c["monetary"]=cltv_c["monetary"]/cltv_c["frequency"]

cltv_c.columns=["recency_w","T_w","frequency","monetary_avg"]

cltv_c
```

# **MODELO CLTV PREDICTION (Valor Futuro del Cliente)**

**SUBMODEL BETAGEOFITTER**

```python
#Aplicamos el Modelo BetaGeoFitter que permite modelar la probabilidad de que el cliente compre en el futuro

from lifetimes import BetaGeoFitter

bgf=BetaGeoFitter(penalizer_coef=0.001) #Aplicamos un coef de penalizacion de 0.001 para evitar valores extremos y mejorar la estabilidad del ajuste.

bgf.fit(cltv_c["frequency"],cltv_c["recency_w"],cltv_c["T_w"])




```

```python
#Top 10 de clientes que esperamos que mas compren una semana desde su ultima compra
bgf.conditional_expected_number_of_purchases_up_to_time(1,cltv_c["frequency"],cltv_c["recency_w"],cltv_c["T_w"]).sort_values(ascending=False).head(10)
```

```python
#Agregamos a la table la Prediccion de compras semanal y semestral
cltv_c["expected_purc_1_week"]=bgf.conditional_expected_number_of_purchases_up_to_time(1,cltv_c["frequency"],cltv_c["recency_w"],cltv_c["T_w"]) #Predice 7 dias desde la ultima compra
cltv_c["expected_purc_6_month"]=bgf.predict(4*6,cltv_c["frequency"],cltv_c["recency_w"],cltv_c["T_w"]) #Predice en lo proximos 6 meses(Pero no desde la ultima compra)

#Le cambio el formato
#cltv_c["expected_purc_1_week"] = cltv_c["expected_purc_1_week"].apply(lambda x: round(x, 6))
#cltv_c["expected_purc_6_month"] = cltv_c["expected_purc_6_month"].apply(lambda x: round(x, 7))

cltv_c.head()
```

```python
#Ventas esperadas en los proximos 6 meses
cltv_c["expected_purc_6_month"].sum()
```

```python
from lifetimes.plotting import plot_period_transactions
#Graficamos la comparacion entre los valores que muestra el modelo y los valores actuales
plot_period_transactions(bgf)
plt.title('Gráfico de Frecuencia de Compras por Período')
plt.xlabel('Número de compras en el período')
plt.ylabel('Cantidad de clientes')
plt.show()
```

**SUBMODEL GAMMA-GAMMA**

```python
from lifetimes import GammaGammaFitter

#Aplicamos el Modelo Gamma-Gamma que permite estimar el valor monetario promedio de las compras de cada cliente

ggf=GammaGammaFitter(penalizer_coef=0.01)
ggf.fit(cltv_c["frequency"],cltv_c["monetary_avg"])
```

```python
#Top 10 de clientes que esperamos que mas ganacias generen
ggf.conditional_expected_average_profit(cltv_c["frequency"],cltv_c["monetary_avg"]).sort_values(ascending=False).head(10)
```

```python
cltv_c["expected_average_profit"]=ggf.conditional_expected_average_profit(cltv_c["frequency"],cltv_c["monetary_avg"])
cltv_c.head()
```
